# **Analyze dataset**

```python
import pandas as pd

# Membaca dataset
weather_data = pd.read_csv('cuaca.csv')

# Informasi dasar tentang dataset
weather_data.info()

cuaca = pd.read_csv('cuaca.csv')
cuaca

# Mengecek nilai yang hilang dan jumlah kategori unik
missing_values = weather_data.isnull().sum()
unique_values = weather_data.nunique()

missing_values, unique_values
```

# **Encoding Data**

1. Menggunakan LabelEncoder untuk mengonversi nilai kategori menjadi angka.
2. Membuat mapping untuk mengonversi nilai kembali ke kategori asli jika diperlukan.

```python
from sklearn.preprocessing import LabelEncoder

# Encoding data ke bentuk numerik
encoder = LabelEncoder()
encoded_data = weather_data.apply(encoder.fit_transform)

# Membuat mapping untuk mengembalikan data ke bentuk asli
inverse_mapping = {col: dict(zip(encoded_data[col], weather_data[col])) for col in weather_data.columns}
```

# **MAKE A NAIVES BAYES MODEL**

```python
from sklearn.naive_bayes import GaussianNB

# Memisahkan fitur dan target
X = encoded_data.drop(columns='Play')
y = encoded_data['Play']

# Melatih model Naive Bayes
nb_model = GaussianNB()
nb_model.fit(X, y)
```

# **PREDICT A SAMPLE**

1. Membuat dataframe dari sampel.
2. Meng-encode kategori sampel ke bentuk numerik menggunakan mapping yang telah dibuat.
3. Menggunakan model untuk memprediksi hasil.
4. Mengonversi hasil prediksi kembali ke kategori asli.

```python
# Mendefinisikan sampel
samples = pd.DataFrame({
    'Outlook': ['Sunny', 'Sunny', 'Rain', 'Overcast', 'Overcast'],
    'Temperature': ['Mild', 'Hot', 'Hot', 'Hot', 'Hot'],
    'Humidity': ['High', 'Normal', 'High', 'Normal', 'Normal'],
    'Windy': ['Strong', 'Strong', 'Strong', 'Weak', 'Strong']
})

# Encode sampel
encoded_samples = samples.apply(lambda col: col.map({v: k for k, v in inverse_mapping[col.name].items()}))

# Prediksi menggunakan model
predictions = nb_model.predict(encoded_samples)

# Decode prediksi ke kategori asli
decoded_predictions = [inverse_mapping['Play'][pred] for pred in predictions]

# Menambahkan hasil prediksi ke tabel sampel
samples['Play(Yes/No)'] = decoded_predictions

print(samples)
```
